```python
# Data analysis Function Declarations

# ----------------------------------------------------------------------------  
#Plotting the time series

# Time series plot of the Closing prices
def ts_plots():

    plt.figure(figsize=(10, 6))
    plt.plot(df['Exchange Date'], df['Close'])
    plt.title('Closing Price Over Time')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.grid(True)
    plt.show()

# Cross-sectional distribution of the Closing prices

    plt.figure(figsize=(10, 6))
    plt.hist(df['Close'], bins=50, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title('Distribution of Closing Prices')
    plt.xlabel('Closing Price')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()


# Time series plot of the daily returns or the differenced Closing price series

    plt.figure(figsize=(10, 6))
    plt.plot(df['Exchange Date'], df['Returns'], color='green')
    plt.title('Daily Returns Over Time')
    plt.xlabel('Date')
    plt.ylabel('Daily Returns')
    plt.grid(True)
    plt.show()

# Cross-sectional distribution of the Daily Returns 

    plt.figure(figsize=(10, 6))
    plt.hist(df['Returns'], bins=35, color='green', edgecolor='black', alpha=0.7)
    plt.title('Distribution of Daily Returns')
    plt.xlabel('Daily Returns')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

# Time series plot of the daily returns or the differenced Closing price series

    plt.figure(figsize=(10, 6))
    plt.plot(df['Exchange Date'], df['Net'], color='orange')
    plt.title('Daily Returns Over Time')
    plt.xlabel('Date')
    plt.ylabel('Net Profit/Loss')
    plt.grid(True)
    plt.show()

# Cross-sectional distribution of the Daily Returns 

    plt.figure(figsize=(10, 6))
    plt.hist(df['Net'], bins=35, color='orange', edgecolor='black', alpha=0.7)
    plt.title('Distribution of Daily Returns')
    plt.xlabel('Net Profit/Loss')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

# ----------------------------------------------------------------------------    
# Plotting the distribution of the daily returns against a normal Distribution 
# with the parameters mean and std obtained from the dataframe

def plot_norm_dist():
    mean_return = df['Returns'].mean() # 0.000026
    std_return = df['Returns'].std() # 0.009910

# Plot the distribution of daily returns using Seaborn
    plt.figure(figsize=(10, 6))
    sns.histplot(df['Returns'], kde=True, color='green', stat='density', bins=35)

# Overlay normal distribution PDF on the plot
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 1000)
    p = norm.pdf(x, mean_return, std_return)
    plt.plot(x, p, 'k', linewidth=2, label='Normal Density')

# Set plot title and labels
    plt.title('Distribution of Daily Returns with Normal PDF overlayed')
    plt.xlabel('Daily Return')
    plt.ylabel('Density')
    plt.legend()

# Show plot
    plt.grid(True)
    plt.show()
    
# ----------------------------------------------------------------------------      
# Plotting the recent regime time series, the moving average and trailing standard deviation


def recent_data_plots():

# Choose the window size for the moving average
    window = 30
    moving_avg = df['Close'].rolling(window=window).mean()
    stdev = df['Close'].rolling(window=window).std()

# Plot the time series of the closing price
    plt.figure(figsize=(10, 6))
    plt.plot(df['Exchange Date'], df['Close'], label='Closing Price')

# Plot moving average of order 30
    plt.plot(df['Exchange Date'], moving_avg, color='red', linestyle='-', label=f'{window}-Day Moving Average')

# Plot standard deviation of order 30
    plt.plot(df['Exchange Date'], moving_avg+stdev, color='green', linestyle='-', label=f'{window}-Standard Deviation')
    plt.plot(df['Exchange Date'], moving_avg-stdev, color='green', linestyle='-', label=f'{window}-Standard Deviation')

    plt.title('Time Series of Closing Price with Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.grid(True)
    plt.show()

###########################

# Plotting a subselection of the Closing Prices time series in its positive-trending regime.

    tseries2=df.iloc[600:940,:]

# Plot of the subselected differenced series, or the Daily Returns

    window_size = 30
    moving_average = tseries2['Returns'].rolling(window=window_size).mean()

    plt.figure(figsize=(10, 6))
    plt.plot(tseries2['Exchange Date'], tseries2['Returns'], color='green')
    plt.plot(tseries2['Exchange Date'], moving_average, color='red')
    plt.title('Daily Returns Over Time')
    plt.xlabel('Date')
    plt.ylabel('Daily Returns')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot of the subselected Closing Prices time series

    window=30
    moving_avg2 = tseries2['Close'].rolling(window=window).mean()

    plt.figure(figsize=(10, 6))
    plt.plot(tseries2['Exchange Date'], tseries2['Close'], label='Closing Price')
    plt.plot(tseries2['Exchange Date'], moving_avg2, color='red', linestyle='-', label=f'{window}-Day Moving Average')
    plt.title('Time Series of Closing Price with Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    
# ----------------------------------------------------------------------------  
# Computing and plotting ACF and PACF values and significance levels

def autocorr_func_plot():

    returns = df['Returns'].dropna()

# Calculate autocorrelation function (ACF) and get p-values
    acf_analysis = acf(returns, alpha=0.05)
    acf_values=acf_analysis[0]
    acf_pvalues=acf_analysis[1]
    acf_values=acf_analysis[0]
    acf_pvalues=acf_analysis[1]

# Calculate partial autocorrelation function (PACF) and get p-values
    pacf_analysis = pacf(returns, alpha=0.05)
    pacf_values=pacf_analysis[0]
    pacf_pvalues=pacf_analysis[1]

    print("acf_values: \n", acf_values, "\n")
    print("pacf_values: \n", pacf_values)

# ACF and PACF plot of the daily returns and check the actual p-values

# ACF plot

    plt.stem(range(len(acf_values)), acf_values)
    plt.title('Autocorrelation Function (ACF)')
    plt.xlabel('Lag')
    plt.ylabel('ACF Value')
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.axhline(y=-1.96/np.sqrt(len(returns)), color='gray', linestyle='--')
    plt.axhline(y=1.96/np.sqrt(len(returns)), color='gray', linestyle='--')
    plt.grid(True)
    plt.show()

# PACF plot

    plt.stem(range(len(pacf_values)), pacf_values)
    plt.title('Partial Autocorrelation Function (PACF)')
    plt.xlabel('Lag')
    plt.ylabel('PACF Value')
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.axhline(y=-1.96/np.sqrt(len(returns)), color='gray', linestyle='--')
    plt.axhline(y=1.96/np.sqrt(len(returns)), color='gray', linestyle='--')
    plt.grid(True)
    plt.show()

# significance threshold

    y1=1.96/np.sqrt(len(returns))
    print(y1)

    for i in range(len(acf_values)):
        if (abs(acf_values[i])>y1):
            print("lag: ", i, " | acf_value: ",acf_values[i])
        
    print("\n---------\n")
        
    for j in range(len(pacf_values)):
        if (abs(pacf_values[j])>y1):
            print("lag: ", j, " | pacf_value: ", pacf_values[j])
```

```python
# Model Functions Declarations


# Computing Closing prices based on a forecasted return series
# The function requires 2 parameters when called, an intial value for the lag[0] used to compute future prices and a forecasted Returns series
def forecast_close_prices(pred0, forecasted_rets):
    forecasted_close=np.array([])
    for j in forecasted_rets:
        pred0=pred0+pred0*j
        forecasted_close=np.append(forecasted_close,pred0)
    return forecasted_close

# Computing the RMSE between 2 series
# The function requires 2 parameters when called, any forecasted series and a real data series
# If the 2 series do not have the same length an error is returned; I have used this as a constraint when I was building and testing the function but all the functions are used correctly this measure is not necessary

def fct_RMSE(forecasted_values,actual_values):
    if len(forecasted_values) != len(actual_values):
        return "Error: Lengths of input Series are not equal", "length of forecast= ", len(forecasted_values),"length of actual values= ", len(actual_values)
    forecasted_values=np.array([forecasted_values])
    actual_values=np.array([actual_values])
    mse = ((actual_values - forecasted_values) ** 2).mean()
    rmse = np.sqrt(mse)
    return rmse

def close_model_vect(df1,a,b,p,d,q,f,log_ret_col=True):

    if log_ret_col:
        retcol='Log_Returns'
    else: 
        retcol='Returns'
        
    df1.index.freq = 'D'
    df1.reset_index(drop=True, inplace=True)
    
    subsection_train=df.loc[a:b]
    subsection_test=df.iloc[b+1:b+f+1]
    
    model = ARIMA(subsection_train[retcol], order=(p,d,q))
    results = model.fit()
    summary=results.summary()
    bic = results.bic
    aic = results.aic
    hqic = results.hqic
    log_likelihood = results.llf
    all_coefs = np.array(results.params)
    p_values = np.array(results.pvalues)
    forecasted_returns = np.array(results.forecast(steps=f))
    
    #Initializing the last lag, used to compute forecasted closing prices based on forecasted returns
    pred=subsection_train['Close'].iloc[-1]
    
    #Computation of the Predicted Closing Prices based on the forcasted returns
    forecasted_close=forecast_close_prices(pred,forecasted_returns)
        
    #Computation of the RMSE of the predicted Returns vs. actual Returns
    rmse_returns=fct_RMSE(forecasted_returns, subsection_test[retcol])

    #Computation of the RMSE of the predicted Closing Prices vs. actual Closing Prices
    rmse_close=fct_RMSE(forecasted_close,subsection_test['Close'])

    
    return summary, "Coefficients: ",all_coefs,"P-values: ", p_values,"Forecasted Returns: ", forecasted_returns,"Returns RMSE: ", rmse_returns, "Forecasted Closing Prices: ",forecasted_close, "Closing Prices RMSE: ", rmse_close 
```

```python
# Function call example using Log returns

close_output=close_model_vect(df,750,934,1,0,3,5,True)
print(close_output[5],close_output[6],close_output[7],close_output[8])
print(close_output[9],close_output[10],close_output[11],close_output[12])
```

```python
# Function call example using Simple Daily Returns

close_output=close_model_vect(df,750,934,2,0,4,5,False)
print(close_output[5],close_output[6],close_output[7],close_output[8])
print(close_output[9],close_output[10],close_output[11],close_output[12])
```

```python
# The function close_model_vect takes 6 positional arguments:
# -df: the dataframe that is being analysed
# -i: the starting point of the lookback period
# -the ending point of the lookback period(in this case 920)
# -the following 3 are the ARIMA(p,d,q) parameters; in this case d is set to 0 as the series is already differentiated, 
#                                                                                                    j is p and k is q
# -the timeframe of the forecast; in this case it is 5 because that is the requirement of the assignment

# RMSE analysis

rmse_array=np.empty((0,4))

for i in range(750,935):
    for j in 1,2,3,4,14,20,24:
        for k in 1,2,3,4,15,20,24:
            x=close_model_vect(df,i,935,j,0,k,5, False)
            rmse_array=np.append(rmse_array, [[i,j,k,x[12]]], axis=0)
```
